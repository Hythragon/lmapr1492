```python
Gamma_X = [0, 21]
X_W = [22, 33]
Gamma_L = [66, 84]
Direction_names = ["Gamma_X", "X_W", "Gamma_L"]
directions = [Gamma_X, X_W, Gamma_L]
for i in range(3):
    ylength = np.linalg.norm(np.array(ph_bs.as_dict()["qpoints"][directions[i][1]]) -np.array(ph_bs.as_dict()["qpoints"][directions[i][0]]))
    for j in range(3):
        v_g = abs(  ph_bs.bands[j][directions[i][0]]- ph_bs.bands[j][directions[i][1]] )/ylength
        print(f"La vitesse du son pour la bande {j}, dans la direction {Direction_names[i]}, est de {v_g*1e12/1e10} m/s")

```
